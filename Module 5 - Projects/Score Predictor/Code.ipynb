```python
import pandas as pd  # type: ignore
import numpy as np  # type: ignore
import matplotlib.pyplot as plt  # type: ignore
```

```python
# Load dataset and prepare data
df = pd.read_csv("score.csv")
df
```

```python
# graph
df.plot(x="Hours", y="Scores", style="o")
plt.title("Hours VS Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Scored")
plt.show
```

```python
# spliting x and y
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values  # Target column as 1D array
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split  # type: ignore
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Train the linear regression model
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Define a function to ask the user for input and predict the score
def predict_score():
    try:
        # Assuming the input feature is a single value (e.g., number of study hours)
        user_input = float(input("Enter the value for the feature (e.g., study hours): "))
        
        # Make sure the input is in the correct shape (2D array) for prediction
        user_data = np.array([[user_input]])
        
        # Predict the score using the trained model
        predicted_score = reg.predict(user_data)
        
        print(f"The predicted score for the input {user_input} is: {predicted_score[0]}")
    
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
```

```python
# Call the function
predict_score()
```

```python
from sklearn import metrics  

y_pred = reg.predict(x_test)
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
